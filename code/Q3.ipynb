```python
import numpy as np
import mat73
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = mat73.loadmat("../data/points2D_Set1.mat")
data_final = [data['x'], data['y']]
data_final = np.array(data_final)

X = data_final[0,:]
Y = data_final[1,:]
```

```python
mean_x = X.sum()/len(X)
mean_y = Y.sum()/len(Y)
mean = np.array([mean_x, mean_y])
mean=mean.reshape(2,1)

Centralized_X = X - mean_x
Centralized_Y = Y - mean_y
```

```python
XY = np.array([Centralized_X, Centralized_Y])
```

```python
var_x = ((X-mean_x)**2).sum()/len(X)
var_xy = ((X-mean_x)*(Y-mean_y)).sum()/len(X)
var_y = ((Y-mean_y)**2).sum()/len(Y)

cov = np.array([[var_x, var_xy],
        [var_xy, var_y]])
```

```python
values, vectors = np.linalg.eig(cov)

pmv = vectors[:, values.argmax()]


plt.scatter(x=X,y=Y,s=5)
plt.title("2D distribution and 1st Principal mode of Variation")
plt.axline(xy1=mean.reshape(2,), slope=pmv[1]/pmv[0], color='r')
plt.axis('equal')
plt.show()
```
